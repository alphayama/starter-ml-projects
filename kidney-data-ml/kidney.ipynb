```python
import xgboost as xgb
import pandas as pd
import numpy as np 
from sklearn import datasets
from sklearn.model_selection import train_test_split,GridSearchCV
from matplotlib import pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import Imputer,StandardScaler,OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
import xgboost as xgb
```

```python
kd=pd.read_csv('./chronic_kidney_with_class.csv')
```

```python
kd.head()
```

```python
kd.replace('?',np.nan,inplace=True)
```

```python
kd.isnull().sum()
```

```python
kd.drop('Unnamed: 0',inplace=True,axis=1)
kd.replace('?',np.nan,inplace=True)
```

```python
kd_dat=kd.drop('class',axis=1)
kd_tgt=kd['class']
```

```python
kd_dat.dtypes
```

```python
kd_dat.dtypes
```

```python
kd_dat[['age','bp','sg','al','su','pot','hemo','pcv','wc','rc','rbc','pc','pcc','ba']]=kd_dat[['age','bp','sg','al','su','pot','hemo','pcv','wc','rc','rbc','pc','pcc','ba']].astype('float')
kd_dat=pd.get_dummies(kd_dat,drop_first=True)
```

```python
kd_dat.head()
```

```python
kd_tgt=pd.get_dummies(kd_tgt,drop_first=True)
```

```python
kd_tgt.head()
```

```python
dtrain,dtest,ltrain,ltest=train_test_split(kd_dat,kd_tgt,test_size=0.3,random_state=29)
```

# Logistic Regression

```python
steps=[
    ('imp',Imputer(missing_values=np.nan,axis=0,strategy='mean')),
    ('scale',StandardScaler()),
    ('log',LogisticRegression())
]
pipeline=Pipeline(steps)
pipeline.fit(dtrain,ltrain)
pipeline.score(dtest,ltest)
```

```python
steps=[
    ('imp',Imputer(missing_values=np.nan,axis=0,strategy='median')),
    ('scale',StandardScaler()),
    ('log',LogisticRegression())
]
pipeline=Pipeline(steps)
pipeline.fit(dtrain,ltrain)
pipeline.score(dtest,ltest)
```

```python
steps=[
    ('imp',Imputer(missing_values=np.nan,axis=0,strategy='most_frequent')),
    ('scale',StandardScaler()),
    ('log',LogisticRegression())
]
pipeline=Pipeline(steps)
pipeline.fit(dtrain,ltrain)
pipeline.score(dtest,ltest)
```

# KNN

```python
steps=[
    ('imp',Imputer(missing_values=np.nan,axis=0,strategy='mean')),
    ('scale',StandardScaler()),
    ('knn',KNeighborsClassifier(n_neighbors=5))
    ]
pipeline=Pipeline(steps)
pipeline.fit(dtrain,ltrain)
pipeline.score(dtest,ltest)
```

# Decision Tree

```python
steps=[
    ('imp',Imputer(missing_values=np.nan,axis=0,strategy='most_frequent')),
    ('scale',StandardScaler()),
    ('dtree',DecisionTreeClassifier(max_depth=20,min_samples_split=0.1,random_state=29))
    ]
pipeline=Pipeline(steps)
pipeline.fit(dtrain,ltrain)
pipeline.score(dtest,ltest)
```

```python
steps=[
    ('imp',Imputer(missing_values=np.nan,axis=0,strategy='mean')),
    ('scale',StandardScaler()),
    ('dtree',DecisionTreeClassifier(max_depth=20,min_samples_split=0.1,random_state=29))
    ]
pipeline=Pipeline(steps)
pipeline.fit(dtrain,ltrain)
pipeline.score(dtest,ltest)
```

```python
steps=[
    ('imp',Imputer(missing_values=np.nan,axis=0,strategy='median')),
    ('scale',StandardScaler()),
    ('dtree',DecisionTreeClassifier(max_depth=20,min_samples_split=0.1,random_state=29))
    ]
pipeline=Pipeline(steps)
pipeline.fit(dtrain,ltrain)
pipeline.score(dtest,ltest)
```

# SVM

```python
steps=[
    ('imp',Imputer(missing_values=np.nan,axis=0,strategy='mean')),
    ('scale',StandardScaler()),
    ('log',SVC())
]
pipeline=Pipeline(steps)
pipeline.fit(dtrain,ltrain)
pipeline.score(dtest,ltest)
```

# XGBoost

```python
steps=[
    ('gscv',xgb.XGBClassifier())
]

param_grid_g={
        'gscv__max_depth':[1,2,5,10,20],
        'gscv__learning_rate':[0.1,0.001,0.3,0.003,0.0001,0.0003],
        'gscv__n_estimators':[5,10,20,30,40]
    }

xgbpipe=Pipeline(steps)
#for i in range(2,20,2):
estim=GridSearchCV(xgbpipe,param_grid=param_grid_g,cv=3,verbose=1,scoring='neg_mean_squared_error')
estim.fit(dtrain,ltrain)
print(estim.best_params_)
print(np.sqrt(-1*estim.best_score_))

```
